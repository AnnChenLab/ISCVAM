```python
import json

# Paths to your existing config file
frontend_config_path = '/uufs/chpc.utah.edu/common/home/u6057318/Documents/ISCVAM_thanh_update/backend/datasets.json'

# Load the existing config file
with open(frontend_config_path, 'r') as f:
    data = json.load(f)

# Lists to store base names for pipeline and second stage datasets
pipeline_dataset_names = {dataset['name'].replace('.pipeline', '') for dataset in data if 'pipeline' in dataset['name']}
secondstage_dataset_names = {dataset['name'].replace('.secondStage', '') for dataset in data if 'secondStage' in dataset['name']}

# Identify pipeline datasets that have corresponding secondStage datasets
to_delete_pipeline = {f"{name}.pipeline" for name in pipeline_dataset_names if name in secondstage_dataset_names}
to_delete_secondstage = {f"{name}.secondStage" for name in secondstage_dataset_names}

# Filter out datasets with names in `to_delete_pipeline` and `to_delete_secondstage`
data = [dataset for dataset in data if dataset['name'] not in to_delete_pipeline and dataset['name'] not in to_delete_secondstage]

# Count datasets with and without "secondStage" in their name for reporting purposes
second_stage_count = len(to_delete_secondstage)
pipeline_deleted_count = len(to_delete_pipeline)
non_second_stage_count = len(data)

# Print the counts
print("Count of datasets with 'secondStage' that were deleted:", second_stage_count)
print("Count of 'pipeline' datasets with matching 'secondStage' that were deleted:", pipeline_deleted_count)
print("Count of remaining datasets without 'secondStage' or matching 'pipeline':", non_second_stage_count)

# Save the updated data back to the same file
with open(frontend_config_path, 'w') as f:
    json.dump(data, f, indent=2)

print("Updated data saved successfully.")
```

```python
import json

# Paths to your existing config file
frontend_config_path = '/uufs/chpc.utah.edu/common/home/u6057318/Documents/ISCVAM_thanh_update/frontend/public/app-settings.json'

# Load the existing config file
with open(frontend_config_path, 'r') as f:
    data = json.load(f)

# Lists to store dataset names for pipeline and second stage
pipeline_dataset = []
secondstage_dataset = []

# Populate lists with dataset names, excluding ".pipeline" and ".secondStage" suffixes
pipeline_dataset.extend(dataset['name'].replace('.pipeline', '') for dataset in data['datasets'] if 'pipeline' in dataset['name'])
secondstage_dataset.extend(dataset['name'].replace('.secondStage', '') for dataset in data['datasets'] if 'secondStage' in dataset['name'])

# Find pipeline datasets that have corresponding secondStage datasets
to_delete = {f"{name}.pipeline" for name in pipeline_dataset if name in secondstage_dataset}

# Delete datasets with "pipeline" in their name if a corresponding "secondStage" dataset exists
data['datasets'] = [dataset for dataset in data['datasets'] if dataset['name'] not in to_delete]

# Save the updated data back to the same file
with open(frontend_config_path, 'w') as f:
    json.dump(data, f, indent=2)

print("Updated data saved successfully.")
print("Deleted pipeline datasets that have corresponding secondStage datasets.")
```
